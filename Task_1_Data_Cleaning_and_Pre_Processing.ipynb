```python
import pandas as pd

# Load dataset
df = pd.read_csv("/content/netflix_titles.csv")

# Step 1: Rename columns to lowercase with underscores
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Step 2: Handle missing values
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['rating'].fillna(df['rating'].mode()[0], inplace=True)
df['duration'].fillna('Unknown', inplace=True)
df = df[df['date_added'].notnull()]  # Drop rows with null date_added

# Step 3: Remove duplicates
df.drop_duplicates(inplace=True)

# Step 4: Convert date_added to datetime and reformat to dd-mm-yyyy
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df['date_added'] = df['date_added'].dt.strftime('%d-%m-%Y')

# Step 5: Standardize text fields
df['country'] = df['country'].str.strip().str.title()
df['rating'] = df['rating'].str.strip().str.upper()
df['type'] = df['type'].str.strip().str.title()

# Step 6: Check and fix data types
df['release_year'] = df['release_year'].astype(int)

# Step 7: Display cleaned data info and sample output
print("Cleaned Dataset Info:\n")
print(df.info())
print("\nSample Rows:\n")
print(df.head())
```
